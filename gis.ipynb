```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cresults = pd.read_csv('results_c.txt')
javaresults = pd.read_csv('results_java.txt')
```

```python
cresults.groupby(['v'])
```

```python
ax =cresults[(cresults.s == 2) & (cresults.v == 1000)].groupby(['e']).mean()[['t']].plot()
javaresults[(javaresults.s == 2) & (javaresults.v == 1000)].groupby(['e']).mean()[['t']].plot(ax=ax)
ax.legend(["C", "Java"])
ax.set_ylabel("time (us)")
ax.set_xlabel("num edges")
ax.set_title("1000 verticles graph")
```

```python
ax = cresults[(cresults['s'] == 2)].groupby(['v']).mean()[['t']].plot(logy=True);
javaresults[(javaresults['s'] == 2)].groupby(['v'])[['t']].mean().plot(ax=ax)
ax.legend(["C", "Java"])
```

```python
for v in [1000, 5000, 10000]:
    for k in [20, 4, 2]:
        print(cresults[(cresults['s'] == 2) & (cresults['v'] == v) & (cresults['e'] == ((v)*(v-1)/k))][['t']].mean())
```

```python
for v in [1000, 5000, 10000]:
    for k in [20, 4, 2]:
        print(javaresults[(javaresults['s'] == 2) & (javaresults['v'] == v) & (javaresults['e'] == ((v)*(v-1)/k))][['t']].mean())
```
